```python
##import dependencies

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from sqlalchemy import create_engine
import psycopg2

from config import user, password
```

```python
engine = create_engine(f'postgresql://{user}:{password}@localhost:5432/employees_db')
connection = engine.connect()

```

# Average Salary by Title

```python
#collect titles df

titles_df = pd.read_sql('SELECT * FROM titles', connection)
titles_df
```

```python
#collect salaries df

salaries_df = pd.read_sql('SELECT * FROM salaries', connection)
salaries_df
```

```python
### link the salaries df to the title df 

employees_df= pd.read_sql('SELECT * FROM employees', connection)
employees_df

merge_df = employees_df.merge(salaries_df, on = "emp_no", how = "outer")
merge_df = merge_df.merge(titles_df, left_on = "emp_title", right_on= 'title_id', how = "outer")
merge_df
```

```python
### generate average salary by title data

avg_salary= merge_df.groupby(by = 'title', dropna = False).mean()
avg_salary
```

```python
#### Bar Chart

fig, ax1 = plt.subplots(figsize=(9, 7))

x_axis = np.arange(len(titles_df))
x_labels = list(avg_salary.index.values)

plt.bar(x_axis, avg_salary['salary'], align= 'center')
plt.xticks(x_axis, x_labels, rotation =90)

plt.title("Average Salary by Title")

plt.ylabel("Average Salary")
plt.xlabeL("Employee Title")

plt.savefig('output/BarPlot_avgsal_title.png')
plt.tight_layout()
plt.show()
```
